```python
students=["vicky","naina","vaishu"]
print(students)
```

```python
message=f"keep it up,{students[0].title()}"
print(message)
```

```python
message=f"keep it up,{students[2].title()}"
print(message)
```

```python
for x in students:
    print(f"well done keep it up,{x.title()}")
```

```python
for y in students:
    print(f"excellent keep more!{y.upper()}")
    print(f"i will be looking forward to receive your github url for the review,{y.title()}")
```

```python
for y in students:
    print(f"excellent keep more!{y.upper()}")
    print(f"i will be looking forward to receive your github url for the review,{y.title()}.\n")
```

```python
for y in students:
    print(f"excellent keep more!{y.upper()}")
    print(f"i will be looking forward to receive your github url for the review,{y.title()}.\n")
print("Thank you for learning")
```
